Running k-means clustering algorithm using Scikit-learn on the Iris dataset. 
Creating silhouette¬†plots for different k values and find the best k.

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
# Compute the Silhouette Coefficient for each sample.
from sklearn.metrics import silhouette_samples
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
import matplotlib.cm as cm
import seaborn as sns
```

#### Load data

```python
iris = load_iris()
X = iris.data
y = iris.target
```

#### Sklearn KMeans

```python
def cluster(X, k=3):
    kmeans_model = KMeans(n_clusters=k).fit(X)
    pred = kmeans_model.predict(X)
    return kmeans_model, pred
```

```python
kmeans_model, y_pred = cluster(X)
```

#### Silhouette Plots
Silhouette analysis can be used to study the separation distance between the resulting clusters. The silhouette plot displays a measure of how close each point in one cluster is to points in the neighboring clusters and thus provides a way to assess parameters like number of clusters visually. This measure has a range of [-1, 1].
The silhouette analysis is used to choose an optimal value for n_clusters.  This is particularly useful for determining cluster imbalance, or for selecting a value for ùêæ by comparing multiple visualizers.

```python
range_n_clusters = [2, 3, 4, 5, 6]
for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, ax1 = plt.subplots(1,1)
    fig.set_size_inches(13, 6)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    n_cluster_kmeans_model = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels_n = n_cluster_kmeans_model.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels_n)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels_n)
    
    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels_n == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.tab20b(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

        ax1.set_title("The silhouette plot for the various clusters.")
        ax1.set_xlabel("The silhouette coefficient values")
        ax1.set_ylabel("Cluster label")

        # The vertical line for average silhouette score of all the values
        ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

        ax1.set_yticks([])  # Clear the yaxis labels / ticks
        ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        
        # We won't plot the clusters like the example did since our data is 4D
        
    plt.show()
```

#### Silhouette Score Analysis
Clustering models with a high Silhouette Coefficient are dense, where samples in the same cluster are similar to each other, and well separated, where samples in different clusters are not very similar to each other.

We can use our silhouette analysis to choose an optimal value for n_clusters. The silhouette plots show that the n_clusters value of 4 and 5 are a bad pick for the given data due to the presence of clusters with below average silhouette scores and also due to wide fluctuations in the size of the silhouette plots. Silhouette analysis is more ambivalent in deciding between 2 and 3.

Also from the thickness of the silhouette plot the cluster size can be visualized. The silhouette plot for cluster 1 when n_clusters is equal to 2, is bigger in size. However when the n_clusters is equal to 6, all the plots are more or less of similar thickness.

 <b>I would tend to go with 3 as an optimal choice of clusters just going off the silhouette plots</b> . I may have chosen k=2 if I had no intuition about the data since it has the highest silhouette coefficient, and since the average silhouette coefficient over the dataset is a measure of how tightly grouped the data are in their clusters and we would like s(i) close to 1. A greater silhouette coefficient corresponds to higher quality clustering.. Further verification of the clusters as 2-D plots by reducing the dimension from 4 to 2 with PCA may be useful, as well as checking the elbow and/or gap methods.

#### Examining Elbow and Gap methods for choosing best K
Scikit-learn do not have a built-in implementation of the gap statistic for k-means clustering. Please find an example of elbow method and gap statistic implementation in this post: https://datasciencelab.wordpress.com/2013/12/27/finding-the-k-in-k-means-clustering/

#### Gap method

* Helper functions

```python
def Wk(mu, clusters):
    K = len(mu)
    return sum([np.linalg.norm(mu[i]-c)**2/(2*len(c)) \
               for i in range(K) for c in clusters[i]])

def bounding_box(X):
    xmin, xmax = min(X,key=lambda a:a[0])[0], max(X,key=lambda a:a[0])[0]
    ymin, ymax = min(X,key=lambda a:a[1])[1], max(X,key=lambda a:a[1])[1]
    return (xmin,xmax), (ymin,ymax)

def find_centers(X, K):
    # Initialize to K random centers
    oldmu = random.sample(list(X), K)
    mu = random.sample(list(X), K)
    while not has_converged(mu, oldmu):
        oldmu = mu
        # Assign all points in X to clusters
        clusters = cluster_points(X, mu)
        # Reevaluate centers
        mu = reevaluate_centers(oldmu, clusters)
    return(mu, clusters)

def has_converged(mu, oldmu):
    return (set([tuple(a) for a in mu]) == set([tuple(a) for a in oldmu]))

def cluster_points(X, mu):
    clusters  = {}
    for x in X:
        bestmukey = min([(i[0], np.linalg.norm(x-mu[i[0]])) \
                    for i in enumerate(mu)], key=lambda t:t[1])[0]
        try:
            clusters[bestmukey].append(x)
        except KeyError:
            clusters[bestmukey] = [x]
    return clusters
 
def reevaluate_centers(mu, clusters):
    newmu = []
    keys = sorted(clusters.keys())
    for k in keys:
        newmu.append(np.mean(clusters[k], axis = 0))
    return newmu

def init_board_gauss(N, k):
    n = float(N)/k
    X = []
    for i in range(k):
        c = (random.uniform(-1, 1), random.uniform(-1, 1))
        s = random.uniform(0.05,0.5)
        x = []
        while len(x) < n:
            a, b = np.array([np.random.normal(c[0], s), np.random.normal(c[1], s)])
            # Continue drawing points from the distribution in the range [-1,1]
            if abs(a) < 1 and abs(b) < 1:
                x.append([a,b])
        X.extend(x)
    X = np.array(X)[:N]
    return X
```

```python
from numpy import zeros
import random
def gap_statistic(X):
    (xmin,xmax), (ymin,ymax) = bounding_box(X)
    # Dispersion for real distribution
    ks = range(1,10)
    Wks = zeros(len(ks))
    Wkbs = zeros(len(ks))
    sk = zeros(len(ks))
    for indk, k in enumerate(ks):
        mu, clusters = find_centers(X,k)
        Wks[indk] = np.log(Wk(mu, clusters))
        # Create B reference datasets
        B = 10
        BWkbs = zeros(B)
        for i in range(B):
            Xb = []
            for n in range(len(X)):
                Xb.append([random.uniform(xmin,xmax),
                          random.uniform(ymin,ymax)])
            Xb = np.array(Xb)
            mu, clusters = find_centers(Xb,k)
            BWkbs[i] = np.log(Wk(mu, clusters))
        Wkbs[indk] = sum(BWkbs)/B
        sk[indk] = np.sqrt(sum((BWkbs-Wkbs[indk])**2)/B)
    sk = sk*np.sqrt(1+1/B)
    return(ks, Wks, Wkbs, sk)
```

```python
# produce an ensemble of 200 data points normally distributed around 3 centers and run the gap statistic on them.
X = init_board_gauss(200,3) 
ks, logWks, logWkbs, sk = gap_statistic(X)
gapnk = logWkbs - logWks
gapnk_diff = [gapnk[i] - gapnk[i+1] for i in range(len(gapnk)-1)]
```

```python
plt.figure(figsize=(16,5))
_ = plt.plot(ks, logWks, color="pink", label="sum of log(W_k)")
_ = plt.plot(ks, logWkbs, label="sum of log(W_kb)")
_ = plt.scatter(ks, gapnk, label="sum of [log(W_kb) - log(W_kb)]")
_ = plt.scatter(ks[:-1], gapnk_diff, label="difference of sum of [log(W_kb) - log(W_kb)]", color='red')
plt.title("K vs sum of log(W_k) and sum of log(W_kb)")
plt.xlabel("K")
plt.ylabel("Sum of log(W)'s'")
plt.legend()
_ = plt.suptitle("Gap method results:")
```

<b>We verified our choice of k=3 with the gap method. If we choose the first k such that gap(k) - gap(k+1) <0,  (seen in red) we would be choosing k=3. </b>

#### Hierarchical Clustering

With hierarchical clustering, elements group together in a tree diagram, and we can choose the end number of clusters after we‚Äôve created the tree. We can use hierarchical clustering two different ways:
* Agglomerative clustering (bottom-up approach): Each data point is a cluster, and we group clusters based on similarity.
* Divisive clustering (top-down approach): The entire data set is a cluster that is split recursively until only singleton clusters of individual data points remain.

a. Look at the hierarchical clustering documentation for [scipy](https://docs.scipy.org/doc/scipy/reference/cluster.hierarchy.html). Perform hierachical clustering on the Iris dataset using single, complete, average and centroid linkage. Plot their associated dendrogram.

```python
linkage_type = ['single', 'complete', 'average', 'centroid']

for lkg in linkage_type:
    # Perform hierarchical/agglomerative clustering
    # linkage(y[, method, metric, optimal_ordering])
    linked = linkage(X, method=lkg) # Here we have the hierarchical clustering represented as a linkage matrix
    plt.figure(figsize=(15, 7))
    dendrogram(linked,
                orientation='top',
                distance_sort='descending',
                show_leaf_counts=True)
    plt.title("Dendrogram for iris data with " + lkg + " linkage")
    plt.show()
```

b. Look at the dendrograms more closely and explain the difference between each linkage method.

Here we have chosen to look into 4 linkage methods:  single, complete, average and centroid linkage. 
##### Single (Minimal) Linkage
This is when we define the cluster dissimilarity metric by the length of the path between the two <b>closest</b> points from each cluster.
Looking at the dendrogram for the hierarchical clustering with single linkage, we can see that there are two big splits in the data. The data here is split about 40/60 between the two main branches. Under that, however, the data seems more spread out, with minimal sub clusters at the high level. 
##### Complete (Maximal) Linkage
This is when we define the cluster dissimilarity metric by the length of the path between the two <b>farthest</b> points from each cluster.
Looking at the dendrogram for the hierarchical clustering with complete linkage, we can see that there are two big splits in the data as well. However, we see that there are clear sub-clusters under each of the two main clusters, and there are even well-defined and even sub-sub-clusters as well. 
##### Average Linkage
This is when we define the cluster dissimilarity metric by the <b>average</b> of the lengths of the paths between <b>each</b> element of cluster 1 and <b>each</b> element of cluster 2.
Looking at the dendrogram for the hierarchical clustering with average linkage, we can see that there are again two big splits in the data. The data here is split unevenly in the two main branches, with about 30% being in one branch and 70% being in another branch. Similar to maximal linkage, we have clearly defined sub-clusters at a high level, but dissimilar to maximal linkage, the sub-sub-clusters have many more branches.  
##### Centroid Linkage
This is when we define the cluster dissimilarity metric by the length of the path between the <b>centroids</b> (multi-dimensional average) of the clusters.
Lastly, looking at the dendrogram for the hierarchical clustering with single linkage, we can see that there are two big splits in the data as well. The two largest splits here do not split the data evenly like the complete linkage did. The two main splits are uneven and the subclusters below have many branches and quickly branch out into sub-sub-clusters. 

c. Choose your linkage method and the number of clusters you would like to keep. Create clustering labels for the Iris dataset and compare with the actual labels. Explain the difference.

Obtain the linkage matrix!

```python
#  the linkage() function will use that method and metric to calculate the distances of the 
# clusters (starting with your n individual samples as singleton clusters)) and in each iteration 
# will merge the two clusters which have the smallest distance according the selected method and metric
linked = linkage(X, method="complete")
```

Retrieve the clusters!

```python
# Form flat clusters from the hierarchical clustering defined by the given linkage matrix
k=3
clusters = fcluster(linked, k, criterion='maxclust')
# We have the clustering labels for the data set, now we can compare to the original labels, y
df_hierarchical = pd.concat([pd.DataFrame(clusters, columns=["Cluster label"]), 
                             dfy], axis=1)
# with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
#     display(df_hierarchical)
```

Compare the clusters to the original labels

```python
species = ["Setosa", "Versicolour", "Virginica"]
print("Cluster distribution per label:\n")
for g, s in zip(df_hierarchical[["Label", "Cluster label"]].groupby("Label"), species):
    print(g[1]["Cluster label"].value_counts())
    plt.figure()
    _ = g[1]["Cluster label"].plot(kind="hist", figsize=(15,4))
    _ = plt.title("Label " + str(g[0]) + " ("+s+" Species) hierarchical cluster distribution")
```

The different linkage methods provide drastically different results in the clustering of our data points. Choosing the 'complete' linkage gives the most similar results to the non-hierarchical clustering we did earlier.
